```python
import pandas as pd
```

```python
df1 = pd.DataFrame([[1,25,'F'],[2,26,'M'],[3,24,'F'],[4,27,'F'],[7,23,'F'],[8,20,'F'],[9,29,'M'],[10,21,'M']],columns=['ID','Age','Gender'])
df1
```

```python
df2 = pd.DataFrame([[1,'Albin'],[2,'Binu'],[3,'Chris'],[4,'Dibin'],[5,'Elbin'],[6,'Febin'],[7,'Gayathri'],[8,'Hari']],columns=['ID','Name'])
df2
```

```python
joinedData = pd.merge(df1,df2,how ="inner",on = "ID")
joinedData
```

```python
joinedData = pd.merge(df1,df2,how = "outer",on = "ID")
joinedData
```

```python
joinedData = pd.merge(df1,df2,how ="left",on = "ID")
joinedData
```

```python
joinedData = pd.merge(df1,df2,how ="right",on = "ID")
joinedData
```

```python
#column level
pd.concat([df1,df2],axis = 1)
```

```python
pd.concat([df1,df2],axis=0)
```

```python
df1 = pd.DataFrame([[1,25,'F'],[2,26,'M'],[3,24,'F'],[4,27,'F'],[7,23,'F'],[8,20,'F'],[9,29,'M'],[10,21,'M']],columns=['ID','Age','Gender'])
df1
```

```python
df2 = pd.DataFrame([[1,'Albin'],[2,'Binu'],[3,'Chris'],[4,'Dibin'],[5,'Elbin'],[6,'Febin'],[7,'Gayathri'],[8,'Hari']],columns=['uniqueID','Name'])
df2
```

```python
pd.merge(df1,df2,how="inner",left_on= "ID",right_on="uniqueID")
```

```python
pd.merge(df1,df2,how="outer",left_on= "ID",right_on="uniqueID")
```
