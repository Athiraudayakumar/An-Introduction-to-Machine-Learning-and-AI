```python
salesData['Total'] = salesData['Total'].apply(lambda x:x.replace('"',''))
```

```python
salesData.dtypes
```

```python
salesData['Total']= salesData['Total'].astype('float64')
```

```python
salesData['Date'] = salesData['Date'].astype('datetime64')
```

```python
salesData.dtypes
```

```python
salesData.head()
```

```python
#1 Total number of customers visited?
totalCustomer = salesData['Invoice ID'].shape[0]
totalCustomer
```

```python
#2 Most visited branch?
branchMode = salesData['Branch'].mode()[0]
branchMode
```

```python
#3  City vs customer analysis? -average spend
filterData = salesData.groupby('City').mean()['Total']
filterData
```

```python
#4 Customer ratio based on customer type?
dataFilter = salesData.groupby(['Customer type']).count()['Invoice ID']/salesData.shape[0]
print(dataFilter)
```

```python
#5 Most sold product line? general analysis
mostSoldProduct = salesData.groupby('Product line').sum()['Quantity']
maxVal = mostSoldProduct.idxmax()
maxVal
```

```python
#6 Costliest product in the shop?
costliestProduct = salesData.groupby('Product line').max()['Unit price']
print(costliestProduct)
costliestItem = costliestProduct.idxmax()
costliestItem
```

```python
#7 Total gross income?(profit)
profit = salesData['Total'].sum()
profit
```

```python
#8 Total gross income per branch?
profitByBranch = salesData.groupby('Branch').sum()['Total']
profitByBranch
```

```python
#9 Total gross income by customer Type?
profitByCustomerType = salesData.groupby('Customer type').sum()['Total']
profitByCustomerType
```

```python
#10 Total gross income by Payment type ?
profitByPaymentType = salesData.groupby('Payment').sum()['Total']
profitByPaymentType
```

```python
#11 Total Margin?
totalGrossMarginePercentage = salesData['gross margin percentage'].sum()
totalGrossMarginePercentage
```

```python
#12 Customer satisfaction analysis based on rating

def satisfaction(rating):
    if rating >= 9:
        rat = 'Highly satisfied'
    elif rating >= 6:
        rat = 'Satisfied'
    else:
        rat = 'disatisfied'
    return rat
```

```python
salesData['Satisfaction'] = salesData['Rating'].map(lambda x:satisfaction(x))
salesData.head()
```

```python
ratingAnalysis = salesData['Satisfaction'].value_counts()
ratingAnalysis
```

```python
ratingAnalysis = salesData[['City','Satisfaction']].value_counts()
ratingAnalysis
```

```python
ratingAnalysis = salesData['Rating'].value_counts()
ratingAnalysis
```

```python
salesData['actualDate'] = salesData['Date'].astype('datetime64')
```

```python
salesData['month'] = salesData['actualDate'].apply(lambda x: x.strftime("%B"))
```

```python
# Monthly income, monthly foot fall - this can be useful for identifying any seasonality
salesPerMonth = salesData.groupby('month').sum()['gross income'].sort_values()
salesPerMonth
```

```python
# Hour is extrating from Time
salesData['hours'] = salesData['Time'].astype('datetime64').apply(lambda x : x.strftime("%H"))
salesData.head()
```

```python
# Hour Based - rush hours? in hour basis
hourBased = salesData.groupby('hours').sum()['gross income'].sort_values()
hourBased
```
