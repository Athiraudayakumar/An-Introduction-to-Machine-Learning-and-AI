# Join


```python
df1 = [[1,25,'F'],[2,26,'M'],[3,24,'F'],[4,27,'F'],[7,23,'M'],[8,20,'F'],[9,29,'M'],[10,21,'M']]
```

```python
df2 = [[1,'Abin'],[2,'Binu'],[3,'Chiris'],[4,'Dibin'],[5,'Ebon'],[6,'febin'],[7,'gayathri']]
data1 = pd.DataFrame(df1,columns=['Id','Age','Gender'])
data1
data2= pd.DataFrame(df2,columns=['Id','name'])
data2
```

```python
combine = pd.concat([data1,data2],axis = 1)
combine
```

```python
mergeData = pd.merge(data1,data2,how ="left",on ="Id")
mergeData
```

```python
mergeData2 = pd.merge(data1,data2,how='inner',on='Id')
mergeData2
```

```python
sep = [['a',23,'mtech'],['b',23,'mtech'],['c',23,'mtech'],['d',23,'mtech'],['e',23,'mtech']]
september = pd.DataFrame(sep,columns=['name','age','education'])
nov = [['a1',23,'mtech'],['b1',23,'mtech'],['c1',23,'mtech'],['d1',23,'mtech'],['e1',23,'mtech']]
november=pd.DataFrame(nov,columns=['name','age','education'])
result = pd.concat([september,november])
result

dataJoined = pd.merge(september,november,how="left",on="name")
dataJoined
```
