```python
import pandas as pd
```

```python
adult = pd.read_csv(r"dataset/adult.csv")
adult
```

```python
#Group the data on the basis of workclass and find average age of each category
grpData = adult.groupby('workclass').mean()['age']
grpData
```

```python
grpDataMul = adult.groupby(['workclass','gender']).mean()[['age','hours-per-week']]
grpDataMul
```

```python
grpDataMul = adult.groupby(['gender','race']).max()[['age','hours-per-week']]
grpDataMul
```

```python
grpDataMul = adult.groupby(['gender','race']).min()[['age','hours-per-week']]
grpDataMul
```

```python
grpData1 = adult.groupby('workclass')
filterData = grpData1[['age','hours-per-week']].mean()
filterData
```
